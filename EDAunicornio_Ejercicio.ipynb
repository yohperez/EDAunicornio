```python
import pandas as pd #Importamos pandas para manejar datos
import matplotlib.pyplot as plt #crear graficos
#Cargamos la data
df = pd.read_csv('Unicorn_Companies.csv')
#Exploración inicial
print("-----Primeras Filas------")
print(df.head()) #muestra las primeras 5 filas
print("-----Información General------")
print(df.info()) #Muestra los tipos de datos y si hay valores nulos
#Limpieza rapida
#Quitar el simbolo de $ y las comas en la columna Valuation para que sea numerica
#Ademas manejaremos los sufijos:
#M para Millions y B para billions
def convert_valuation (valuation_str):
  valuation_str = str(valuation_str).replace('$', '').replace(',', '').strip()
  if 'B' in valuation_str:
    return float(valuation_str.replace('B', '')) * 1_000_000_000
  elif 'M' in valuation_str:
      return float(valuation_str.replace('M', '')) * 1_000_000
  else:
      return float(valuation_str)
#Estadística descriptiva
print("------Resumen Estadistico-----")
print(df.describe()) #Muestra media, máximo y minimo etc
df['Industry'].value_counts().plot(kind='bar', title='Sectores con mas unicornios')
plt.ylabel('Cantidad de empresas') #etiqueta o nombre del grafico
plt.show()
```
